```python
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
lmbd = 2
c = 1
a = 1
K = 30
r = 1
iterations_t = 200
```

```python
hosts = []
parasites = []
```

```python
def initialize(hosts, parasites):

    hosts.clear()
    parasites.clear()

    initial_h = (c * np.log(lmbd)) / ((lmbd - 1) * a * c) + 0.3
    hosts.append(initial_h)

    initial_p = (np.log(lmbd) / a) + 0.3
    parasites.append(initial_p)

    print("in init", len(hosts))
    print(initial_p, initial_h)
```

```python
def run_simulation(k=None, rieker=False):

    initialize(hosts, parasites)

    for t in range(1, iterations_t+1):

        print("hosts: ", hosts[t-1])
        print("Iteration ", t)

        
        next_h = lmbd * hosts[t-1] * np.exp(-a * parasites[t-1])

        if rieker:
            
            next_h = hosts[t-1] * np.exp(r * (1 - hosts[t-1] / K) - a * parasites[t-1])

        hosts.append(next_h)

        next_p = c * hosts[t-1] * (1 - np.exp(-a * parasites[t-1]))
        
        if k and next_p < .01:
            next_p = k * parasites[t-1]
            

        parasites.append(next_p)

```

```python
run_simulation()
```

```python
x = np.arange(iterations_t+1)
plt.plot(x, np.log(hosts), label = "hosts")
plt.plot(x, np.log(parasites), label = "parasites")
plt.legend()
plt.show()
```

```python
plt.scatter(np.log(hosts), np.log(parasites))
```
