```python
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
lmbd = 2
c = 1
a = 1
iterations_t = 200
```

```python
hosts = []
parasites = []
```

```python
initial_h = (c * np.log(lmbd)) / ((lmbd - 1) * a * c) + 0.3
hosts.append(initial_h)

initial_p = (np.log(lmbd) / a) + 0.3
parasites.append(initial_p)

print(initial_p, initial_h)
```

```python
def run_simulation(k=None):
    for t in range(1, iterations_t+1):

        print("Iteration ", t)
        hosts.append(
            lmbd * hosts[t-1] * np.exp(-a * parasites[t-1])
        )

        next_p = c * hosts[t-1] * (1 - np.exp(-a * parasites[t-1]))
        
        if k:
            next_p += k * hosts[t-1]

        parasites.append(next_p)

```

```python
run_simulation()
```

```python
x = np.arange(iterations_t+1)
plt.plot(x, np.log(hosts), label = "hosts")
plt.plot(x, np.log(parasites), label = "parasites")
plt.legend()
plt.show()
```

```python
plt.scatter(np.log(hosts), np.log(parasites))
```

```python
k = .1
run_simulation(k)
```
